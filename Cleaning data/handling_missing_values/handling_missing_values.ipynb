# Handling Missing Values

```python
import pandas as pd
import numpy as np
```

```python
# read in all our data
nfl_data = pd.read_csv(r'C:\Users\CHUWI\OneDrive\Máy tính\python MCI\DA Kaggle\CLeanning_data\NFL Play by Play 2009-2017 (v4).csv')

```

```python
nfl_data.head()
```

### How many missing data points do we have?

```python
# get number of missing data points per column.
missing_values_count = nfl_data.isnull().sum()

# look at the missing points in the first ten columns
missing_values_count[0:10]
```

##### it's seem like a lot!
##### check for missing data by percentage

```python
# how many total missing values do we have?
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells)*100
print(' we have about {} percent chance of missing data'.format(percent_missing))
```

## Figure out why the data is missing
##### i will find Is this value missing because it wasn't recorded or because it doesn't exist?

```python
missing_values_count[0:10]
```

```python
# remove all columns with at least one missing values
column_with_na_dropped = nfl_data.dropna(axis=1)
column_with_na_dropped.head()
```

```python
# just how much data did we lose?
print('Columns in original dataset: ',nfl_data.shape[1])

print("Columns with na's dropped",column_with_na_dropped.shape[1])
```

### Filling in missing values automatically

```python
subset_nfl_data = nfl_data.loc[:,"EPA":"Season"].head()
subset_nfl_data
```

#### replace missing values with whatever value comes directly after it in the same column. (This makes a lot of sense for datasets where the observations have some sort of logical order to them.)

```python
# Replace all NA's the value that comes directly after it in the same column, 
# Then replace all the remaining na's with 0
subset_nfl_data.fillna(method="bfill",axis=1).fillna(0)
```
