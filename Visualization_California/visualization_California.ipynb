```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities_df = pd.read_csv('./california_cities.csv')
```

```python
cities_df.head()

```

```python
# Extract Latd,lon 
lat,lon = cities_df['latd'], cities_df['longd']
population, area = cities_df['population_total'], cities_df['area_total_km2']

plt.style.use('seaborn')
plt.figure(figsize = (8,6))

# plot uwsing Pyplot API
plt.scatter(lon,lat,
           c= np.log10(population), cmap='viridis',
           s = area, linewidths = 0, alpha = 0.5)
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)

#Create a legend for cities sizes
area_range =  [50,100,150,200,300,500]
for area in area_range:
    plt.scatter([],[], s= area,c='k',alpha = 0.4,label=str(area)+ 'km$^2$')

plt.legend(labelspacing = 1, title = 'city Area')

plt.title('California Cities: Population and Area Distribution')
```
